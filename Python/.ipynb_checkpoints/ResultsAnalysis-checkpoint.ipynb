```python
import os, pandas as pd, numpy as np, matplotlib.pyplot as plt
```

```python
baseDir = 'C:\\Users\\mtcraig\\Desktop\\Research\\Models\\MacroCEM\\Python'
os.chdir(baseDir)

#dirs = ['Results_WECC_NetZero_DACS2050_reference_TrueREFERENCE']#,'Results_WECC_NetZero_DACS2050_reference_TrueREFERENCE_Hydro']
dirs = os.listdir('Results')
print(dirs)

dirs = ['Results_EI_NetZero_DACS2050_reference_TrueREFERENCE']

yrs = list(range(2022,2041,2))

transEff = .95
stoFTLabels = ['Energy Storage','Pumped Storage']
```

# CHECK RESULTS

```python
def sumVarToRegion(df,fleet):
    dfT = df.T
    dfT['region'] = fleet['region']
    return dfT.groupby('region').sum().T

def sumVarToFT(df,fleet):
    dfT = df.T
    dfT['FuelType'] = fleet['FuelType']
    return dfT.groupby('FuelType').sum().T

def sumFlowsToRegion(flows,lines):
    flowsT = flows.T
    flowsT['r'] = lines['r']
    flowsT['rr'] = lines['rr']
    return flowsT.groupby('r').sum().T,flowsT.groupby('rr').sum().T
```

```python
for d in dirs:
    for yr in yrs:
        print(d,yr)
        yr = str(yr)
#         os.chdir(os.path.join(baseDir,'Results',d,yr+'CO2Cap0','CE'))
        os.chdir(os.path.join(baseDir,d,yr+'CO2Cap0','CE'))
        
        #Import vars
        demand = pd.read_csv('demandCE' + str(yr) + '.csv',header=0,index_col=0,parse_dates=True)
        gen = pd.read_csv('vGenCE' + str(yr) + '.csv',header=0,index_col=0,parse_dates=True)*1000
        gentech = pd.read_csv('vGentechCE' + str(yr) + '.csv',header=0,index_col=0,parse_dates=True)*1000
        flows = pd.read_csv('vLineflowCE' + str(yr) + '.csv',header=0,index_col=0,parse_dates=True)*1000
        charge = pd.read_csv('vChargeCE' + str(yr) + '.csv',header=0,index_col=0,parse_dates=True)*1000
        chargetech = pd.read_csv('vChargetechCE' + str(yr) + '.csv',header=0,index_col=0,parse_dates=True)*1000
        
        #Import inputs
        fleet = pd.read_csv('genFleetForCE' + str(yr) + '.csv',header=0,index_col=0)
        fleet.index = fleet['GAMS Symbol']
        techs = pd.read_csv('newTechsCE' + str(yr) + '.csv',header=0,index_col=0)
        techs.index = techs['GAMS Symbol']
        lines = pd.read_csv('lineLimitsAfterCE' + str(yr) + '.csv',header=0,index_col=0)
        lines.index = lines['GAMS Symbol']
        
        #Map data to regions
        genRegion = sumVarToRegion(gen,fleet)
        chargeRegion = sumVarToRegion(charge,fleet)
        genTechRegion = sumVarToRegion(gentech,techs)
        chargeTechRegion = sumVarToRegion(chargetech,techs)
        flowOut,flowIn = sumFlowsToRegion(flows,lines)
        
        #Account for transmission efficiency
        flowIn *= transEff
        
##############Supply and demand balance
        totalDemand = demand + chargeRegion + chargeTechRegion + flowOut
        totalSupply = genRegion + genTechRegion + flowIn
        
        print('Supply and demand balance:')
        print((totalDemand-totalSupply).max().max())
        print((totalDemand-totalSupply).min().min())

        #Examine times when supply >>> demand
        if (totalDemand-totalSupply).min().min() < -1e3:
            plt.figure()
            totalDemand.reset_index(inplace=True,drop=True),totalSupply.reset_index(inplace=True,drop=True)
            (totalDemand-totalSupply).plot()
            plt.title(d+str(yr))
            
            l = list()
            for n,df in zip(['demand','charge','charge tech','out','gen','gen tech','in'],
                             [demand,chargeRegion,chargeTechRegion,flowOut,genRegion,genTechRegion,flowIn]):
                s = df.sum(axis=1)
                s.name = n
                l.append(s)
            df = pd.concat(l,axis=1)
#             df.to_csv('hourlySDBalance.csv')
            
            plt.figure()
            df.reset_index(inplace=True,drop=True)
            df.plot()
            plt.title(d+str(yr))
            
            minHr = (totalDemand-totalSupply).sum(axis=1).idxmin()
            plt.figure()
            df.loc[minHr-3:minHr+3].plot()  
            
            genFT = sumVarToFT(gen,fleet)
            gentechFT = sumVarToFT(gentech,techs)
            genFT = genFT.add(gentechFT,fill_value=0)
            genFT.reset_index(inplace=True,drop=True)
            genFT = genFT.loc[minHr]
            plt.figure()
            genFT.plot.bar()
        
##############Min/max generation constraints
        print('Generation min/max:')
        print(gen[gen<0].sum().sum())
        print((gen.max(axis=0)-fleet['Capacity (MW)']).max())
        
        #Max transmission flows
        print('Flows:')
        print((flows.max(axis=0)-lines['TotalCapacity']).max())

##############Storage operations
        soc = pd.read_csv('vStateofchargeCE' + str(yr) + '.csv',index_col=0,header=0,parse_dates=True)*1000
        sto = fleet.loc[fleet['FuelType'].isin(stoFTLabels)]
        stoGen,stoCharge,soc = gen.T.loc[sto.index],charge.T.loc[sto.index],soc.T.loc[sto.index]
        print('Storage:')
        print((stoGen.T.max()-sto['Capacity (MW)']).max())
        print((stoCharge.T.max()-sto['Capacity (MW)']).max())
        print((soc.T.max()-sto['Nameplate Energy Capacity (MWh)']).max())

##############New storage operations
        vE = pd.read_csv('vEneBuiltSto' + str(yr) + '.csv',index_col=0,header=0)*1000
        vC = pd.read_csv('vPowBuiltSto' + str(yr) + '.csv',index_col=0,header=0)*1000
        socTech = pd.read_csv('vStateofchargetechCE' + str(yr) + '.csv',index_col=0,header=0,parse_dates=True)*1000
        sto = techs.loc[techs['FuelType'].isin(stoFTLabels)]
        stoGenTech,stoChargeTech,socTech = gentech.T.loc[sto.index],chargetech.T.loc[sto.index],socTech.T.loc[sto.index]
        print('New storage:')
        print((stoGenTech.T.max()-vC['0']).max())
        print((stoChargeTech.T.max()-vC['0']).max())
        print((socTech.T.max()-vE['0']).max())
                
##############Hydropower operations
        print('Hydropower:')
        hydro = fleet.loc[fleet['FuelType']=='Hydro']
        blocks = pd.read_csv('hoursCEByBlock' + str(yr) + '.csv',index_col=0,header=0,parse_dates=True)
        hydroMaxGen = pd.read_csv('hydroGenCE' + str(yr) + '.csv',index_col=0,header=0)
        if hydro.shape[0] > 0:
            genT = gen.T
            genT = genT.loc[hydro.index]
            genT.index = hydro['region']
            hydroGen = genT.T
            for b in blocks['block'].unique():
                g = hydroGen.loc[blocks.loc[blocks['block']==b].index]
                g = g.sum(axis=0)
                g.name = b
                maxGen = hydroMaxGen.loc[b]
                print((g-maxGen).max())
```

# ANALYZE INVESTMENTS

```python
genCap,lineCap,stoP,stoE = list(),list(),list(),list()
for d in dirs:
    for yr in yrs:
        os.chdir(os.path.join(baseDir,d,str(yr)+'CO2Cap0','CE'))
        
        #Get fleet, filtering out retired units
        fleet = pd.read_csv('genFleetAfterCE' + str(yr) + '.csv',header=0,index_col=0)
        fleet.index = fleet['GAMS Symbol']
        fleet = fleet.loc[fleet['Retired']==False]
        
        ftCapacity = fleet.groupby('FuelType')['Capacity (MW)'].sum()
        ftCapacity.name = yr
        genCap.append(ftCapacity)
        
        #Get transmission investments
        lines = pd.read_csv('lineLimitsAfterCE' + str(yr) + '.csv',header=0,index_col=0)
        lines.index = lines['GAMS Symbol']
        
        lines = lines.iloc[:lines.shape[0]//2] #get rid of reflected lines
        lines = lines['TotalCapacity']
        lines.name = yr
        lineCap.append(lines)
        
        #Get storage investments
        e = pd.read_csv('vEneBuiltSto' + str(yr) + '.csv',index_col=0,header=0)
        p = pd.read_csv('vPowBuiltSto' + str(yr) + '.csv',index_col=0,header=0)
        e,p = e['0'],p['0']
        e.name,p.name = yr,yr
        stoP.append(p),stoE.append(e)
        
genCap = pd.concat(genCap,axis=1)
genCap.plot.bar()

plt.figure()
lineCap = pd.concat(lineCap,axis=1)
lineCap.plot.bar()

plt.figure()
p = pd.concat(stoP,axis=1)
p.plot.bar()

plt.figure()
e = pd.concat(stoE,axis=1)
e.plot.bar()
```

# ANALYZE GENERATION

```python
ftGen,regionGen = list(),list()
for d in dirs:
    for yr in yrs:
        os.chdir(os.path.join(baseDir,d,str(yr)+'CO2Cap0','CE'))

        #Import vars
        gen = pd.read_csv('vGenCE' + str(yr) + '.csv',header=0,index_col=0)*1000
        gentech = pd.read_csv('vGentechCE' + str(yr) + '.csv',header=0,index_col=0)*1000
        
        #Import inputs
        fleet = pd.read_csv('genFleetForCE' + str(yr) + '.csv',header=0,index_col=0)
        fleet.index = fleet['GAMS Symbol']
        techs = pd.read_csv('newTechsCE' + str(yr) + '.csv',header=0,index_col=0)
        techs.index = techs['GAMS Symbol']
        
        #Sum data to fuel type
        genFT = sumVarToFT(gen,fleet).sum()
        gentechFT = sumVarToFT(gentech,techs).sum()
        genFT = genFT.add(gentechFT,fill_value=0)
        genFT.name = yr
        ftGen.append(genFT)
        
        #Sum data to region
        genRegion = sumVarToRegion(gen,fleet).sum()
        gentechRegion = sumVarToRegion(gentech,techs).sum()
        genRegion = genRegion.add(gentechRegion,fill_value=0)
        genRegion.name = yr
        regionGen.append(genRegion)
    
gen = pd.concat(ftGen,axis=1)
gen.plot.bar()

# print(gen)

plt.figure()
gen = pd.concat(regionGen,axis=1)
gen.plot.bar()        
```

```python
yrs = [2040]
for d in dirs:
    for yr in yrs:
        #Import inputs
        fleet = pd.read_csv('genFleetForCE' + str(yr) + '.csv',header=0,index_col=0)
        fleet.index = fleet['GAMS Symbol']
        techs = pd.read_csv('newTechsCE' + str(yr) + '.csv',header=0,index_col=0)
        techs.index = techs['GAMS Symbol']

        #Import operational and build variables
        gen = pd.read_csv('vGenCE' + str(yr) + '.csv',header=0,index_col=0,parse_dates=True)*1000
        gentech = pd.read_csv('vGentechCE' + str(yr) + '.csv',header=0,index_col=0,parse_dates=True)*1000
        charge = pd.read_csv('vChargeCE' + str(yr) + '.csv',header=0,index_col=0,parse_dates=True)*1000
        chargetech = pd.read_csv('vChargetechCE' + str(yr) + '.csv',header=0,index_col=0,parse_dates=True)*1000
        soc = pd.read_csv('vStateofchargeCE' + str(yr) + '.csv',index_col=0,header=0,parse_dates=True)*1000
        vE = pd.read_csv('vEneBuiltSto' + str(yr) + '.csv',index_col=0,header=0)*1000
        vC = pd.read_csv('vPowBuiltSto' + str(yr) + '.csv',index_col=0,header=0)*1000
        socTech = pd.read_csv('vStateofchargetechCE' + str(yr) + '.csv',index_col=0,header=0,parse_dates=True)*1000
        
        #Isolate storage units
        sto = fleet.loc[fleet['FuelType'].isin(stoFTLabels)]
        stoGen,stoCharge,soc = gen.T.loc[sto.index],charge.T.loc[sto.index],soc.T.loc[sto.index]

        sto = techs.loc[techs['FuelType'].isin(stoFTLabels)]
        stoGenTech,stoChargeTech,socTech = gentech.T.loc[sto.index],chargetech.T.loc[sto.index],socTech.T.loc[sto.index]

        stoGen,stoCharge,soc = stoGen.T,stoCharge.T,soc.T
        stoGen.reset_index(inplace=True,drop=True),stoCharge.reset_index(inplace=True,drop=True),soc.reset_index(inplace=True,drop=True)
        
        styles = ['b-','r--','y:']
        linewidths = [2,1,4]    
        for g in stoGen.columns:
            df = pd.concat([stoGen[g],stoCharge[g],soc[g]],axis=1)
            df.columns = ['gen','charge','soc']
            fig, ax = plt.subplots()
            for col, style, lw in zip(df.columns, styles, linewidths):
                df[col].plot(style=style, lw=lw, ax=ax)
                ax.legend()
                plt.title(g)
        
        stoGenTech,stoChargeTech,socTech = stoGenTech.T,stoChargeTech.T,socTech.T
        stoGenTech.reset_index(inplace=True,drop=True),stoChargeTech.reset_index(inplace=True,drop=True),socTech.reset_index(inplace=True,drop=True)
        for g in stoGenTech.columns:
            df = pd.concat([stoGenTech[g],stoChargeTech[g],socTech[g]],axis=1)
            df.columns = ['gen','charge','soc']
            fig, ax = plt.subplots()
            for col, style, lw in zip(df.columns, styles, linewidths):
                df[col].plot(style=style, lw=lw, ax=ax)
                ax.legend()
                plt.title(g)
```
